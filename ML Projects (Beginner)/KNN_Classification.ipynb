```python
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier

digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data
```

```python
df = pd.DataFrame(digits.data, digits.target)

df.head()
```

```python
df['target'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3, random_state=10)
```

Creating KNN Classifier

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test,y_test)
```

Manually trying other values of k to find the best one :

```python
knn = KNeighborsClassifier(n_neighbors=8)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=25)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test,y_test)
```

After more trial and error, came to the conclusion that 5 is the best value for k.

```python
from sklearn.metrics import confusion_matrix
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
