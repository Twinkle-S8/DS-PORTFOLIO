Outlier removal using percentile method on AirBnB data of NYC (Dataset from Kaggle)

```python
import pandas as pd
df = pd.read_csv("/AB_NYC_2019.csv.zip")
df.head()
```

```python
df.shape
```

```python
df.price.describe()
```

```python
min_thresold, max_thresold = df.price.quantile([0.01,0.999])
min_thresold, max_thresold
```

```python
df[df.price<min_thresold]
```

Removing outliers based on max & min threshold.

```python
df2 = df[(df.price>min_thresold)&(df.price<max_thresold)]
df2.shape
```

```python
df2.sample(5)
```

```python
df2.price.describe()
```
